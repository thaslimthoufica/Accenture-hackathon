```python
import os
import zipfile

# 📦 Paths
zip_path = "/content/Dataset.zip"
extract_path = "content/Dataset.zip/[Usecase 7] AI-Driven Customer Support Enhancing Efficiency Through Multiagents​\Conversation"  # simpler path
combined_txt_path = "/content/combined_customer_support.txt"

# 1. 🔓 Extract ZIP
os.makedirs(extract_path, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# 2. 📚 Combine all TXT files
combined_text = ""

for root, dirs, files in os.walk(extract_path):
    for file in files:
        if file.endswith(".txt"):
            file_path = os.path.join(root, file)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    combined_text += f.read() + "\n\n"
            except Exception as e:
                print(f"Error reading {file_path}: {e}")

# 3. 💾 Save combined text to file
with open(combined_txt_path, "w", encoding="utf-8") as f:
    f.write(combined_text)

print("✅ Combined text file saved!")

# 4. 🧠 Train the Retrieval-Based QA Model
from langchain_community.document_loaders import TextLoader, CSVLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFaceHub

# 📁 Paths again
csv_path = "/content/specific_data/Dataset/[Usecase 7] AI-Driven Customer Support Enhancing Efficiency Through Multiagents​/Historical_ticket_data.csv"

# Load TXT and CSV
text_loader = TextLoader(combined_txt_path)
text_documents = text_loader.load()

csv_loader = CSVLoader(file_path=csv_path)
csv_documents = csv_loader.load()

# Combine and split
all_documents = text_documents + csv_documents
splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = splitter.split_documents(all_documents)

# Embeddings and Vectorstore
embedding_model = HuggingFaceEmbeddings(model_name="mistralai/Mistral-7B-Instruct-v0.1")
vectorstore = FAISS.from_documents(chunks, embedding_model)

# HuggingFace LLM (make sure API key is set)
llm = HuggingFaceHub(
    repo_id="mistralai/Mistral-7B-Instruct-v0.1",  # or any other
    model_kwargs={"temperature": 0.5, "max_length": 512}
)

# Create the Retrieval QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=vectorstore.as_retriever(),
    return_source_documents=True
)

# ✅ Test it!
query = "How can I reset my password?"
response = qa_chain({"query": query})
print(response["result"])


```
